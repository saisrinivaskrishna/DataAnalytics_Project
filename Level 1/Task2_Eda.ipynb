```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display graphs inside notebook
%matplotlib inline

# Better plot style
sns.set(style="whitegrid")
```

```python
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM",
    "AGE", "DIS", "RAD", "TAX", "PTRATIO",
    "B", "LSTAT", "MEDV"
]

df = pd.read_csv(
    "../datasets/house Prediction Data Set.csv",
    header=None,
    names=column_names,
    delim_whitespace=True
)
```

```python
df.head()
```

```python
print("Shape:", df.shape)

print("\nInformation")
df.info()

print("\nSummary Statistics")
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MEDV"], bins=30, kde=True, color="royalblue")

plt.title("Distribution of House Prices")
plt.xlabel("MEDV")
plt.ylabel("Frequency")
plt.savefig("Histogram.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python

plt.figure(figsize=(12,10))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.savefig("Heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

Observation:
RM has a strong positive correlation with MEDV.
LSTAT has a strong negative correlation with MEDV.
TAX and RAD are highly positively correlated

```python

plt.figure(figsize=(8,5))

sns.boxplot(x=df["MEDV"])

plt.title("Boxplot of House Prices")
plt.savefig("Boxplot.png", dpi=300, bbox_inches="tight")

plt.show()
```

Observation:
A few outliers are present among higher-priced houses.



```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="RM",
    y="MEDV",
    data=df,
    color="green"
)

plt.title("Number of Rooms vs House Price")
plt.savefig("Scatterplot.png", dpi=300, bbox_inches="tight")

plt.show()
```

Observation:Houses with more rooms generally have higher prices.

```python
pair = sns.pairplot(
    df[["RM", "LSTAT", "PTRATIO", "MEDV"]],
    diag_kind="hist"
)

pair.savefig("Pairplot.png", dpi=300)

plt.show()
```

```python

Q1 = df["MEDV"].quantile(0.25)
Q3 = df["MEDV"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["MEDV"] < lower) | (df["MEDV"] > upper)]

print("Number of Outliers:", len(outliers))
```
